```python
from matplotlib import pyplot as plt
from numpy import linspace

#Circle of Radius 2
#Integrating

def f(x):
    return (4 -x**2)**(.5)

a = 0 
b= 2

#How many slices
#10 slices needs 11 points
N=1000
h=(b-a) /(N)
x=linspace(a,b,N+1)
plt.plot(x, f(x))
plt.show()

#Rectangle method
#for i in range(N):
#    s+= (f(x[i+1])
#s*=h
#print(s)

#Start integral at 0
#Trapezoid method 
s=0
for i in range(N):
    s+= (f(x[i+1])+f(x[i]))/2
s*=h
print(s)


```
