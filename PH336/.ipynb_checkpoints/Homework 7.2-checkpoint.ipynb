```python
import numpy as np
from numpy.linalg import inv
import matplotlib.pyplot as plt
from sympy import pprint

#Create array of xdata
xdata=np.linspace(50, 240, 20)

#Create array of ydata
ydata=np.array([5.0,7.0,11.0,13.0,21.0,43.0,30.0,16.0,15.0,10.0,13.0,42.0,90.0,75.0,29.0,13.0,8.0,4.0,6.0,3.0])

print(xdata, ydata)
```

```python
#Define Lorentzian function
def Lorentz(x, μ, Γ):
    return (1/np.pi)*((Γ/2)/((x-μ)**2+(Γ/2)**2))
```

```python
#Define given constants
μ1=102.1
μ2=177.9
Γ1=30.0
Γ2=20.0
σ=np.sqrt(ydata)

#Create X matrix
X=np.array([Lorentz(xdata, μ1, Γ1)/σ, Lorentz(xdata, μ2, Γ2)/σ]).T

#Create Y matrix
Y=ydata/σ
```

```python
#Create ϵ matrix
ϵ=inv(np.matmul(X.T,X))

ϵ

#Create matrix for solved linear constants
a=np.matmul(ϵ,np.matmul(X.T, Y))
#[a1, a2]
print(a)
```

```python
#Create fit line
fit=a[0]*Lorentz(xdata, μ1, Γ1)+ a[1]*Lorentz(xdata, μ2, Γ2)

#Calculate chi-squared of the fit
Χ2=sum((ydata-fit)**2/σ)


#Create Histogram and Fit Line

plt.plot(xdata, ydata, 'bo')
plt.plot(xdata, fit, '-', color='#ff7f50')
plt.show()

print('Chi-squraed= '+ str(Χ2))
print('Amplitude')
print(a)

print('Error Matrix')
pprint(ϵ)
```
